## Insight 1: Prioritize Warm Leads Over Cold Outreach

**Finding:**
- Cold contacts (no prior history) convert at only **8.83%**
- Contacts with previous successful campaigns convert at **65.11%**
- That's a **7.4x difference** in conversion rate

**Recommendation:**
- Build a **lead scoring system** that prioritizes previously
  successful contacts first
- Allocate at least 60% of campaign budget to warm leads
- Use cold outreach only as a supplementary channel

**Expected Impact:**
- Reducing cold outreach by 30% while doubling warm outreach
  could increase overall conversion rate from 11.27% to ~18-22%

## Insight 2: Switch Budget from Telephone to Cellular

**Finding:**
| Channel | Conversion Rate |
|---------|----------------|
| Cellular | 14.74% |
| Telephone | 5.23% |

- Cellular is **2.8x more effective** than telephone
- 15,044 contacts were made via telephone, largely wasted effort

**Recommendation:**
- Shift at least 70% of telephone budget to cellular outreach
- Retire landline telephone campaigns entirely for new contacts
- Use telephone only for elderly segments with no cellular access

**Expected Impact:**
- Converting 30% of telephone contacts to cellular could
  generate ~500-700 additional conversions per campaign cycle

## Insight 3: Stop Over-Contacting: Less is More

**Finding:**
| Contacts Made | Conversion Rate |
|---------------|----------------|
| 1 contact | 13.04% |
| 2-3 contacts | 11.22% |
| 4-5 contacts | 8.68% |
| 6+ contacts | 5.49% |

- Every additional contact **reduces** the chance of conversion
- Contacts made 6+ times convert at less than half the rate of
  single-contact leads

**Recommendation:**
- Set a **hard cap of 3 contacts per lead** per campaign
- After 3 failed contacts, move lead to a nurture email sequence
- Reallocate saved call time to new high-quality leads

**Expected Impact:**
- Eliminating 6+ contact attempts frees up ~3,385 call slots
  that can be used on fresh, higher-converting leads

## Insight 4: Focus on High-Converting Customer Segments

**Finding:**
| Segment | Group | Conversion Rate |
|---------|-------|----------------|
| Age | 65+ | 46.85% |
| Job | Student | 31.43% |
| Job | Retired | 25.23% |
| Age | 17-25 | 22.99% |
| Job | Blue-collar | 6.89% |
| Age | 26-35 | 8.91% |

**Recommendation:**
- Design **targeted campaigns** specifically for 65+ and students
- Create age-appropriate messaging:
  - 65+: security, stability, long-term savings
  - Students: future planning, first savings account
- Reduce investment in blue-collar and 26-35 segments or
  redesign the offer for those segments

**Expected Impact:**
- Reallocating 20% of blue-collar budget to 65+ and student
  segments could yield 2-3x more conversions per contact

## Insight 5: Time Campaigns Around High-Converting Months

**Finding:**
- March converts at **50.55%** highest of any month
- September, October, December also perform strongly
- May has the most contacts but a relatively low conversion rate

**Recommendation:**
- Launch major campaign pushes in **March, September, October**
- Reduce campaign volume in low-converting months
- Investigate what drives March performance
  (economic conditions, bonus season, tax season timing)
- Align campaign calendar with these peaks every year

**Expected Impact:**
- Shifting 15% of May contacts to March could add
  ~300-400 additional conversions with the same budget
