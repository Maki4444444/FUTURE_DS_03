# Step 5: Segmentation Analysis
### Marketing Funnel Analysis: Bank Marketing Campaign Dataset

### Goal
Break down conversion performance across key customer segments to identify
which groups respond best to the marketing campaign.

### Segments We'll Analyze
1. Job type
2. Education level
3. Marital status
4. Age group
5. Previous campaign outcome
6. Month + Day of week

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('../data/bank_cleaned.csv')
print(f"Data loaded: {df.shape[0]:,} rows × {df.shape[1]} columns")
```

```python
def conversion_summary(df, col):
    """Returns conversion rate summary grouped by a column."""
    summary = df.groupby(col, observed=True)['y_binary'].agg(['sum', 'count'])
    summary.columns = ['converted', 'total']
    summary['conversion_rate_%'] = (summary['converted'] / summary['total'] * 100).round(2)
    summary['drop_off_%'] = (100 - summary['conversion_rate_%']).round(2)
    return summary.sort_values('conversion_rate_%', ascending=False)
```

```python
print("=== CONVERSION BY JOB TYPE ===")
job_seg = conversion_summary(df, 'job')
print(job_seg.to_string())
```

```python
print("=== CONVERSION BY EDUCATION LEVEL ===")
edu_seg = conversion_summary(df, 'education')
print(edu_seg.to_string())
```

```python
print("=== CONVERSION BY MARITAL STATUS ===")
marital_seg = conversion_summary(df, 'marital')
print(marital_seg.to_string())
```

```python
print("=== CONVERSION BY AGE GROUP ===")
age_seg = conversion_summary(df, 'age_group')
print(age_seg.to_string())
```

```python
print("=== CONVERSION BY PREVIOUS CAMPAIGN OUTCOME ===")
poutcome_seg = conversion_summary(df, 'poutcome')
print(poutcome_seg.to_string())
```

```python
print("=== CONVERSION BY MONTH ===")
month_order = ['jan','feb','mar','apr','may','jun',
               'jul','aug','sep','oct','nov','dec']
month_seg = conversion_summary(df, 'month')
month_seg = month_seg.reindex([m for m in month_order if m in month_seg.index])
print(month_seg.to_string())
```

```python
print("=== CONVERSION BY DAY OF WEEK ===")
day_order = ['mon', 'tue', 'wed', 'thu', 'fri']
day_seg = conversion_summary(df, 'day_of_week')
day_seg = day_seg.reindex(day_order)
print(day_seg.to_string())
```

```python
print("=== TOP PERFORMING SEGMENTS SUMMARY ===\n")

print(f"Best Job:        {job_seg.index[0]} ({job_seg['conversion_rate_%'].iloc[0]}%)")
print(f"Best Education:  {edu_seg.index[0]} ({edu_seg['conversion_rate_%'].iloc[0]}%)")
print(f"Best Age Group:  {age_seg.index[0]} ({age_seg['conversion_rate_%'].iloc[0]}%)")
print(f"Best Month:      {month_seg['conversion_rate_%'].idxmax()} ({month_seg['conversion_rate_%'].max()}%)")
print(f"Best Day:        {day_seg['conversion_rate_%'].idxmax()} ({day_seg['conversion_rate_%'].max()}%)")
print(f"Best Poutcome:   {poutcome_seg.index[0]} ({poutcome_seg['conversion_rate_%'].iloc[0]}%)")
```

```python
print("=== JOB × CHANNEL CONVERSION ===")
cross = df.groupby(['job', 'contact'], observed=True)['y_binary'].agg(['sum', 'count'])
cross.columns = ['converted', 'total']
cross['conversion_rate_%'] = (cross['converted'] / cross['total'] * 100).round(2)
cross = cross.sort_values('conversion_rate_%', ascending=False)
print(cross.head(10).to_string())
```

```python
job_seg.to_csv('../outputs/segment_job.csv')
edu_seg.to_csv('../outputs/segment_education.csv')
age_seg.to_csv('../outputs/segment_age.csv')
month_seg.to_csv('../outputs/segment_month.csv')
day_seg.to_csv('../outputs/segment_day.csv')
poutcome_seg.to_csv('../outputs/segment_poutcome.csv')

print("All segmentation files saved to outputs/")
```

## Segmentation Summary
- Some job types convert significantly better than others
- Education level shows clear conversion pattern differences
- Previous campaign success is the strongest predictor of conversion
- Month and day of week show seasonal and weekly patterns
- Cross-segment analysis reveals best job × channel combinations
